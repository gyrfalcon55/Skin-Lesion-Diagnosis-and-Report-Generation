```python
import tensorflow as tf
import os

IMAGE_DIR = r"D:\Artificial Intelligence\College project\dataverse_files\skin_lesion_images\data"
MASK_DIR  = r"D:\Artificial Intelligence\College project\dataverse_files\skin_lesion_images\masked_data"

IMG_SIZE = (128, 128)
BATCH_SIZE = 16
```

```python
image_files = sorted(os.listdir(IMAGE_DIR))
mask_files  = sorted(os.listdir(MASK_DIR))
```

```python
def load_image_mask(image_name, mask_name):
    # Load image
    img_path = tf.strings.join([IMAGE_DIR, image_name], separator=os.sep)
    img = tf.io.read_file(img_path)
    img = tf.image.decode_jpeg(img, channels=3)
    img = tf.image.resize(img, IMG_SIZE)
    img = tf.cast(img, tf.float32) / 255.0

    # Load mask
    mask_path = tf.strings.join([MASK_DIR, mask_name], separator=os.sep)
    mask = tf.io.read_file(mask_path)
    mask = tf.image.decode_png(mask, channels=1)
    mask = tf.image.resize(mask, IMG_SIZE, method="nearest")
    mask = tf.cast(mask > 127, tf.float32)

    return img, mask
```

```python
TRAIN_RATIO = 0.8
VAL_RATIO   = 0.2
```

```python
dataset_size = len(image_files)
AUTOTUNE = tf.data.AUTOTUNE
```

```python
dataset = tf.data.Dataset.from_tensor_slices((image_files, mask_files))

dataset = dataset.map(
    lambda x, y: load_image_mask(x, y),
    num_parallel_calls=AUTOTUNE
)

dataset = dataset.shuffle(buffer_size=dataset_size, seed=42, reshuffle_each_iteration=False)
```

```python
train_size = int(TRAIN_RATIO * dataset_size)
val_size   = int(VAL_RATIO * dataset_size)

train_ds = dataset.take(train_size)
val_ds   = dataset.skip(train_size).take(val_size)

train_ds = (
    train_ds
    .shuffle(500)              # reshuffle every epoch
    .batch(16)                  # IMPORTANT
    .prefetch(1)
)

val_ds = (
    val_ds
    .batch(16)
    .prefetch(1)
)
```

```python
print("Train batches:", tf.data.experimental.cardinality(train_ds))
print("Val batches:", tf.data.experimental.cardinality(val_ds))
```

```python
def conv_block(x, filters):
    x = layers.Conv2D(filters, 3, padding='same', activation='relu')(x)
    x = layers.Conv2D(filters, 3, padding='same', activation='relu')(x)
    return x
```

```python
def build_cpu_unet():
    inputs = layers.Input(shape=(128, 128, 3))

    # Encoder
    c1 = conv_block(inputs, 32)
    p1 = layers.MaxPooling2D()(c1)

    c2 = conv_block(p1, 64)
    p2 = layers.MaxPooling2D()(c2)

    c3 = conv_block(p2, 128)
    p3 = layers.MaxPooling2D()(c3)

    # Bottleneck
    bn = conv_block(p3, 256)

    # Decoder
    u1 = layers.UpSampling2D()(bn)
    u1 = layers.Concatenate()([u1, c3])
    c4 = conv_block(u1, 128)

    u2 = layers.UpSampling2D()(c4)
    u2 = layers.Concatenate()([u2, c2])
    c5 = conv_block(u2, 64)

    u3 = layers.UpSampling2D()(c5)
    u3 = layers.Concatenate()([u3, c1])
    c6 = conv_block(u3, 32)

    outputs = layers.Conv2D(1, 1, activation='sigmoid')(c6)

    return Model(inputs, outputs,name="CPU_U-Net")
```

```python
output_channels = 1
model = build_cpu_unet()
optimizer = tf.keras.optimizers.Adam(learning_rate=1e-4)
model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])
```

```python
for img, mask in train_ds.take(1):
    print("Image range:", tf.reduce_min(img).numpy(), tf.reduce_max(img).numpy())
    print("Mask values:", tf.unique(tf.reshape(mask, [-1]))[0].numpy())
```

```python
from tensorflow.keras.callbacks import EarlyStopping

early_stopping_callbacks = EarlyStopping(monitor='val_loss',patience=3,restore_best_weights=True)
```

```python
history = model.fit(train_ds,validation_data=val_ds,epochs=7,callbacks=[early_stopping_callbacks])
```

```python
import tensorflow as tf

unet_model = tf.keras.models.load_model(
    "models/unet_model(128).h5",
    compile=False
)
```

```python
import os
import cv2
import numpy as np

IMG_SIZE = (128, 128)   # or (128,128)  EXACT size used in training
```

```python
def preprocess_for_unet(image_path):
    img = cv2.imread(image_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = cv2.resize(img, IMG_SIZE)
    img = img.astype("float32") / 255.0
    return np.expand_dims(img, axis=0)  # (1, H, W, 3)
```

```python
def predict_mask(image_path):
    x = preprocess_for_unet(image_path)
    pred = unet_model.predict(x, verbose=0)[0]   # (H, W, 1)
    mask = (pred > 0.5).astype("uint8")          # binary
    return mask
```

```python
def create_masked_image(original_path, mask):
    original = cv2.imread(original_path)
    original = cv2.resize(original, IMG_SIZE)

    mask_3ch = np.repeat(mask, 3, axis=-1)
    masked_rgb = original * mask_3ch

    return masked_rgb
```

```python
CLASSIFICATION_DIR = r"D:\Artificial Intelligence\College project\skin_lesion_data\test"
OUTPUT_DIR         = r"D:\Artificial Intelligence\College project\skin_lesion_data\masked_test"

os.makedirs(OUTPUT_DIR, exist_ok=True)

for class_name in os.listdir(CLASSIFICATION_DIR):
    class_path = os.path.join(CLASSIFICATION_DIR, class_name)
    save_class_path = os.path.join(OUTPUT_DIR, class_name)

    if not os.path.isdir(class_path):
        continue

    os.makedirs(save_class_path, exist_ok=True)

    for img_name in os.listdir(class_path):
        img_path = os.path.join(class_path, img_name)

        # 1. predict mask
        mask = predict_mask(img_path)

        # 2. generate masked image
        masked_img = create_masked_image(img_path, mask)

        # 3. save
        save_path = os.path.join(save_class_path, img_name)
        cv2.imwrite(save_path, masked_img)

    print(f"Done class: {class_name}")
```
